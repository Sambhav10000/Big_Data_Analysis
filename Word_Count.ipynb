```python
from pyspark import SparkContext

sc = SparkContext.getOrCreate()

# Specify the path to the input text file
file_path = "/content/Basketball_info.txt"

# Read the file and proceed with the rest of the word count logic
line = sc.textFile(file_path)
word_counts = line.flatMap(lambda line: line.split(" ")).map(lambda word: (word, 1)).reduceByKey(lambda x, y: x + y)

for word, count in word_counts.collect():
    print(f"{word}: {count}")

#Stop the SparkContext
sc.stop()
```
